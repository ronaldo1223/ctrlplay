```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.uol.com.br/start/esport/'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

lista = ['lol','valorant','brawl star','pol','fortnite','free fire']

for paragrafo in soup.find_all('body'):
        for palavra in lista:
            for paragrafo_str in paragrafo.stripped_strings:
                if palavra.upper() in str(paragrafo_str).upper():
                    print('NOTICIA SOBRE: {}\n{}\n'.format(palavra.upper(),paragrafo_str))
```

```python
import requests
from bs4 import BeautifulSoup

from tkinter import *
from tkinter import  messagebox 

class application:

    def __init__(self, master=None):
    
        self.fontpadrao = ('arial','14')
    
        
        self.primeiroconteiner = Frame(master)
        self.primeiroconteiner['pady'] = 10
        self.primeiroconteiner.pack()

        self.segundoconteiner = Frame(master)
        self.segundoconteiner['pady'] = 10
        self.segundoconteiner.pack()

        self.terceiroconteiner = Frame(master)
        self.terceiroconteiner['pady'] = 10
        self.terceiroconteiner.pack()
    
        self.titulo = Label(self.primeiroconteiner, text= 'analisador de sites')
        self.titulo['font'] = ('ariel','26','bold')
        self.titulo.pack()

        self.senhaLabel = Label(self.segundoconteiner, text = 'site: ', font = self.fontpadrao)
        self.senhaLabel.pack(side=LEFT)

        self.senha = Entry(self.segundoconteiner)
        self.senha['width'] =30
        self.senha['font'] = self.fontpadrao
        self.senha.pack(side=LEFT)

        self.autenticar = Button(self.terceiroconteiner)
        self.autenticar ['text'] = 'verificar'
        self.autenticar['font'] = ('calibri','14')
        self.autenticar['width'] = '12'
        self.autenticar['command'] =self.verificaSite
        self.autenticar.pack()


        self.mensagem = Label(self.terceiroconteiner, text = '', font = self.fontpadrao)
        self.mensagem.pack()

    def verificaSite(self):
        try:
            site = self.senha.get()
            resultado = requests.get(site)
            resposta = resultado.text
            soup = BeautifulSoup(resposta, 'html.parser')
        
        except:
            messagebox.showinfo(
            title = 'erro',
            message = 'favor inserir um site valido')
        else:
            opcao = messagebox.askokcancel(title = 'analisador de sites',message = 'site: {} \n pronto para continuar?'.format(site))


            self.mensagem['text'] = soup.title.string
                
                
                                

root = Tk()
application(root)
root.mainloop()
```
